```python
import pandas as pd
# Connect to S3
s3 = session.client("s3")

# Bucket and file
bucket_name = "my-unique-dataset-bucketm"
file_key = "AB_NYC_2019.csv"

# Load CSV into pandas
obj = s3.get_object(Bucket=bucket_name, Key=file_key)
df = pd.read_csv(obj["Body"])

df.head()
```

```python
# --- 2. Handle Missing Values ---
# Impute NaN in 'reviews_per_month' with 0 (since they correspond to 0 reviews)
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# Drop high-cardinality/redundant/administrative columns
columns_to_drop = ['id', 'host_id', 'host_name', 'name', 'last_review', 'neighbourhood']
df = df.drop(columns=columns_to_drop)
```
